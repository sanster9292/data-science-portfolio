```
def result_or(df):
  if df['Industry']=='Advertising' or df['Category'] == "Blue":
    return 'Yes'
  else:
    return 'No'
  
def result_and(df):
  if df['Industry']=='Advertising' and df['Category'] == "Blue":
    return 'Yes'
  else:
    return 'No'  
  
df['Result1 (If-And)'] = df.apply(result_or, axis =1)
df['Result2 (If-And)'] = df.apply(result_and, axis =1)
    
  
```

```
import numpy as np
def formay(x):
  if np.isnan(x) == True:
    return np.nan
  else:
    return "${:.1f}".format((x))

df['Split-Cost1'] = [float(x.split(':')[1].strip('$').replace(',','')) for x in df['Cost']]
# formatters={'Split-Cost1': lambda x: ''if type(x)==str else '\u20B9{:.2f}'.format(x)}
# # display(df2.style.format(formatters))            
# df2['Split-Cost1'] = pd.to_numeric(df2['Split-Cost1'])

df['Split-Cost1'] = df['Split-Cost1'].apply(formay)

df
```
